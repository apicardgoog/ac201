## Problem 1

Given a list of name, generate the corresponding initials using list comprehension. Then determine who has matching initials. 

```python
presidents = "George Washington,John Madden,John Adams,Thomas Jefferson,James Madison,James Monroe,John Quincy Adams,Andrew Jackson,Martin Van Buren,William Henry Harrison,John Tyler,James K. Polk,Zachary Taylor,Millard Fillmore,Franklin Pierce,James Buchanan,Abraham Lincoln,Andrew Johnson,Ulysses S. Grant,Rutherford B. Hayes,James A. Garfield,Chester A. Arthur,Grover Cleveland,Benjamin Harrison,William McKinley,Theodore Roosevelt,William Howard Taft,Woodrow Wilson,Warren G. Harding,Calvin Coolidge,Herbert Hoover,Franklin D. Roosevelt,Harry S. Truman,Dwight D. Eisenhower,John F. Kennedy,Lyndon B. Johnson,Richard Nixon,Gerald Ford,Jimmy Carter,Ronald Reagan,George H. W. Bush,Bill Clinton,George W. Bush,Barack Obama,Donald Trump"
names_list = presidents.split(",")

def GetInitials(name):
    return "".join([x[0] for x in name.split(" ")])

initials_list = [(GetInitials(x), x) for x in names_list]
initials_dict = {}
for initials, name in initials_list:
    if initials not in initials_dict:
        initials_dict[initials] = []
    initials_dict[initials].append(name)

for initials in initials_dict:
    names_list = initials_dict[initials]
    if len(names_list) > 1:
        print("The initials", initials, "are shared by", ", ".join(names_list[:-1]), "and", names_list[-1])

```

What's the country with highest population density?


```python
df[df["Pop. Density"] == max(df["Pop. Density"])]
```

What country has the highest GDP in Oceania?

```python
oceania = df[df["Region"].str.strip() == "OCEANIA"]
oceania[oceania.GDP == max(oceania.GDP)]
```

How many countries are in each region?

```python
df.groupby("Region").count()[["Country"]]
```
